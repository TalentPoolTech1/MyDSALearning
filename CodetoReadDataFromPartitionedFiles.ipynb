```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import dayofmonth, col, input_file_name, month, year
```

```python
ss = SparkSession.builder.appName('Get EachFileSize').master('local[1]').getOrCreate()
```

```python
import os
file_list = os.listdir(path="C:\\Users\\Venkateswarlu.A\\Downloads\\NewYorkTaxiData")
print(os.stat(file_list))
```

```python
df = ss.read.csv("C:\\Users\\Venkateswarlu.A\\Downloads\\NewYorkTaxiData\\PARTITIONED_DATA\\*.csv",inferSchema=True, header=True)
df = df.withColumn("in_file_name", input_file_name())
df = df.withColumn("pickup_year",year(col("tpep_pickup_datetime"))).withColumn("pickup_month", month("tpep_pickup_datetime")).withColumn("pickup_day",dayofmonth(col("tpep_pickup_datetime")))
print(df.show())
print(df.groupBy(["in_file_name","pickup_year","pickup_month","pickup_day"]).count().show())
```
